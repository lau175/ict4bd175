```python
#nameFile='/home/ict4bd/Desktop/Atene/creation_epw/station_data/LGAV_from_2013-01-01_to_2013-12-31.csv'
#nameFile='/home/ict4bd/Desktop/Atene/creation_epw/station_data/LGAV_from_2014-01-01_to_2014-12-31.csv'
nameFile='/home/ict4bd/Desktop/Atene/creation_epw/station_data/LGAV_from_2015-01-01_to_2015-12-31.csv'
meteo_df=pd.read_csv(nameFile,index_col='Date_Time',parse_dates=[0,])
meteo_df
```

```python
meteo=meteo_df.resample('H').mean().interpolate()
meteo.head()
```

```python
if 'Solar_W/m2' in list(meteo.columns):
    meteo=meteo.drop(columns='Solar_W/m2')

if 'UV' in list(meteo.columns):
    meteo=meteo.drop(columns='UV')
```

```python
#retrieve year 
name_csv=nameFile.split("/")[-1]
start_date=name_csv.split("_")[-3]
year=start_date.split("-")[-3]
```

```python
out_dir='/home/ict4bd/Desktop/Atene/creation_epw/'+year+'_meteo_files/'+year+'/'

if not os.path.exists(os.path.dirname(out_dir)):
    try:
        os.makedirs(os.path.dirname(out_dir))
    except OSError as exc: # Guard against race condition
        if exc.errno != errno.EEXIST:
            raise
```

```python
time_vector = pd.date_range(start='1/1/'+str(int(year)), end='1/1/'+str(int(year)+1), freq='H')
newdf = pd.DataFrame(index=time_vector, columns=meteo.columns)
newdf.drop(newdf.tail(1).index,inplace=True) # drop last 1 rows
newdf
```

```python
newdf.loc[meteo.index, meteo.columns] = meteo.loc[meteo.index, meteo.columns]
newdf
```

```python
import time
from datetime import timedelta

df0 = newdf.copy()
mask = newdf.isna()
missing_values_time = df0["Temperature_C"][mask["Temperature_C"]].index

while df0["Temperature_C"].isna().any():
    try:
        df0.loc[missing_values_time] = df0.loc[missing_values_time - timedelta(days=1)].values
    except:
        df0.loc[missing_values_time] = -100000555555

df0
```

```python
df0= df0.replace(-100000555555, np.nan)
df0.to_csv(out_dir+"meteo_"+year)
df0
```

```python
df = df0.copy()
mask = df.isna()
missing_values_time = df["Temperature_C"][mask["Temperature_C"]].index

while df["Temperature_C"].isna().any():
    try:
        df.loc[missing_values_time] = df.loc[missing_values_time + timedelta(days=1)].values
    except:
        break
df
```

```python
df.to_csv(out_dir+'meteo_'+year+'_updated')
meteo.columns
```

```python
#check for difference (leap year)
if len(df)!=len(ghi_tot):
    start=datetime.datetime.strptime(year+'-02-28 23:59:59', '%Y-%m-%d %H:%M:%S')
    end=datetime.datetime.strptime(year+'-02-29 23:59:59', '%Y-%m-%d %H:%M:%S')
    df_365=df[(df.index<start) | (df.index>end)]
    df=df_365.copy()
    print('Leap day: dropped')
    print(len(df))
else:
    print('Not a leap year')
```
